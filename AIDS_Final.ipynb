### 2010-2018年各国艾滋病艾滋病新增人数

```python
from pyecharts import options as opts
from pyecharts.charts import Map
from pyecharts.charts import Bar, Page, Pie, Timeline

def timeline_map() -> Timeline:
    tl = Timeline()
    for i in range(2011,2019):
        map0 = (
            Map()
            .add(
                "艾滋病新增人数", list(zip(list(df.CountryName),list(df["{}".format(i)]))), "world",is_map_symbol_show = False
            )
            .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
            .set_global_opts(
                title_opts=opts.TitleOpts(title="2010-2018年各国艾滋病艾滋病新增人数".format(i),
                                         subtitle_textstyle_opts=opts.TextStyleOpts(color="red",font_size=10,font_style="italic")),
                
                visualmap_opts=opts.VisualMapOpts(series_index=0,max_=100000),
            
            )
        )
        tl.add(map0, "{}".format(i))
    return tl

艾滋病新增人数世界地图 = timeline_map()
艾滋病新增人数世界地图.render_notebook()
```

* ### 我将2010年-2018年的艾滋病新增人数进行了数据可视化，可以看出，除去人口基数问题，每年艾滋病人口增长最多的一直都以是南非，莫桑比克，赞比亚，尼日利亚，津巴布韦等非洲国家为主。

```python
from pyecharts import options as opts
from pyecharts.charts import Bar
def bar_base() -> Bar:
    c = (
        Bar()
        .add_xaxis(x轴)
        .add_yaxis("颁布前", before)
        .add_yaxis("颁布当年", effect)
        .add_yaxis("颁布后", after)
        .set_global_opts(title_opts=opts.TitleOpts(title="颁布法律前，中，后对比", subtitle="同性婚姻合法化国家颁布法律前，中，后艾滋病新增人数"))
    )
    return c

c = bar_base()
c.render_notebook()
```

## 结论

* 同性婚姻合法化不会导致严重的艾滋病人数增加，反之，许多通过同性婚姻合法化的国家艾滋病新增人数有所下降。
* 同性性行为不是艾滋病最主要的因素，更关键的是教育水平，卫生环境，以及个人素养等。
* 中国目前艾滋病病情不算严重，但因为我国人口基数大，艾滋病患者数量多，所以艾情也需要引起重视。
